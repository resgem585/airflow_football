```python
import pandas as pd 
import time
import random
import os
#from utils import get_data,data_processing
from datetime import datetime
import uuid
```

```python
# Crear una lista con las url de cada pais y otra lista del pais al que pertenece
url = ['https://www.espn.com.co/futbol/posiciones/_/liga/esp.1',
       'https://www.espn.com.co/futbol/posiciones/_/liga/eng.1',
       'https://www.espn.com.co/futbol/posiciones/_/liga/ita.1',
       'https://www.espn.com.co/futbol/posiciones/_/liga/ger.1',
       'https://www.espn.com.co/futbol/posiciones/_/liga/fra.1',
       'https://www.espn.com.co/futbol/posiciones/_/liga/por.1',
       'https://www.espn.com.co/futbol/posiciones/_/liga/ned.1']
ligas = ['ESPAÑA', 'INGLATERRA', 'ITALIA', 'ALEMANIA', 'FRANCIA', 'PORTUGAL', 'HOLANDA']

# Creación del DataFrame
df_ligas = {
    'LIGA': ligas,
    'URL': url
} 

df_ligas = pd.DataFrame(df_ligas)
df_ligas
```

```python
# Probamos con la posicion 1 del dataframe 
df_ligas['URL'][1]
```

```python
# Guardar la informacion obtenida de las tablas HTML 
df = pd.read_html(df_ligas['URL'][1])
```

```python
# Verificar la cantidad de registros que nos trajo

len(df)
```

```python
# Vemos la data del primer registro

df[0]
```

```python
# Ver la data sel segundo registro

df[1]
```

```python
# Concatenar los dos registros

df=pd.concat([df[0],df[1]],ignore_index=True,axis=1)
df
```

```python
# Limpieza de la data 

df=df.rename(columns={0:'EQUIPO',1:'J', 2:'G', 3:'E', 4:'P', 5:'GF', 6:'GC', 7:'DIF', 8:'PTS'})
df['EQUIPO']=df['EQUIPO'].apply(lambda x: x[5:] if x[:2].isnumeric()==True else x[4:])
df['LIGA'] = df_ligas['LIGA'][1]
```

```python
df
```

```python
# Crear un CSV que contenga la liga y el url

df_ligas.to_csv('df_ligas.csv',index=False)
```

```python
# Generamos un ID aleatorio 

def id_generator(team):

    return str(uuid.uuid1())[:8]
```

```python
lista_resultado = 0
```

```python
# Generar un id generico para cada equipo 

equipos = df['EQUIPO'].unique()
lista_resultado = list(map(id_generator, equipos))
```

```python
# Crea un diccionario donde la clave 'EQUIPO' contiene los nombres de los equipos y la clave 'ID_TEAM' contiene los IDs correspondientes
team_table = {
    'EQUIPO': equipos,         # La lista equipos contiene los nombres de los equipos
    'ID_TEAM': lista_resultado  # La lista lista_resultado contiene los IDs correspondientes a los equipos
}

# Convierte el diccionario team_table en un DataFrame de pandas
df_team = pd.DataFrame(team_table)
```

```python
# Utiliza la función value_counts() para contar la frecuencia de cada valor único en la columna 'ID_TEAM' del DataFrame df_team
id_counts = df_team['ID_TEAM'].value_counts()

# Muestra el conteo de cada valor único en la columna 'ID_TEAM'
print(id_counts)
```

```python
# Guarda el DataFrame df_team en un archivo CSV llamado 'team_table.csv' sin incluir el índice de fila
df_team.to_csv('team_table.csv', index=False)
```

```python
# Realiza una fusión (merge) interna entre los DataFrames df y df_team, utilizando la columna 'EQUIPO' como clave de fusión
df_final = pd.merge(df, df_team, how='inner', on='EQUIPO')

# Muestra los nombres de las columnas del DataFrame df_final
print(df_final.columns)

# Selecciona y reordena las columnas del DataFrame df_final
df_final = df_final[['ID_TEAM', 'EQUIPO', 'J', 'G', 'E', 'P', 'GF', 'GC', 'DIF', 'PTS', 'LIGA']]
```

```python
# Guarda el DataFrame df_final en un archivo CSV llamado 'premier_positions.csv' sin incluir el índice de fila
df_final.to_csv('premier_positions.csv', index=False)
```
